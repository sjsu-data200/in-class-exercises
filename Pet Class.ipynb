```python
import datetime
from datetime import date
from dataclasses import dataclass
```

# Pet Class I

```python
class Pet:
    def __init__(self, species, breed, name, color, gender, owner, address, birthday):
        self._species = species
        self._breed = breed
        self._name = name
        self._color = color
        self._gender = gender
        self._owner = owner
        self._address = address
        self._birthday = birthday
    
    def getSpecies(self):
        return self._species 

    def getBreed(self):
        return self._breed 
    
    def getName(self):
        return self._name 
    
    def getColor(self):
        return self._color 
    
    def getGender(self):
        return self._gender 

    def getOwner(self):
        return self._owner 
    
    def getAddress(self):
        return self._address 
    
    def getBirthday(self):
        return self._birthday
    
    def setSpecies(self,new_species):
        self.species = new_species
    
    def setBreed(self,new_breed):
        self._breed = new_breed

    def setName(self,new_name):
        self._name = new_name
        
    def setColor(self,new_color):
        self._color = new_color
        
    def setGender(self,new_gender):
        self._gender = new_gender
        
    def setOwner(self,new_owner):
        self._owner = new_owner
        
    def setAddress(self,new_address):
        self._address = new_address
        
    def setBirthday(self,new_birthday):
        if type(new_birthday) == date:
            self._birthday = new_birthday
        
    def getAge(self):
        now = date.today()
        return (now - self._birthday).days//365
    
    def displayGender(self):
        if self._gender == 1:
            print('Male')
        elif self._gender == 0:
            print('Female')
        else:
            print('Unkonwn Gender')
    
    def __str__(self):
        return ("Pet Inforamtion\n----------------------\n{:<15} {}\n{:<15} {} \n{:<15} {}\n{:<15} {}").format('Name',self._name,'Age in Years',self.getAge(),'Species',self._species,'Color',self._color)
    
class Dog(Pet):
    
    def __init__(self, breed, name, color, gender):
        super().__init__('Canis Familiaris', breed, name, color, gender,None,None,None)
           
    def getAge(self):
        if self._birthday == None:
            return 'Unknown'
        now = date.today()
        return (now - self._birthday).days//365   

class Puppy(Dog):
    def __init__(self,breed,name,color,gender):
        super().__init__(breed,name,color,gender)
    
    def getAge(self):
        if self._birthday == None:
            return 'Unknown'
        now = date.today()
        return (now - self._birthday).days//30
    
    def play(self):
        return "(___()'`;\n      /,    /`\n      \\\\\"--\\\\"
       
    def __str__(self):
        return ("Pet Inforamtion\n----------------------\n{:<15} {}\n{:<15} {} \n{:<15} {}\n{:<15} {}").format('Name',self._name,'Age in Months',self.getAge(),'Species',self._species,'Color',self._color)
        
        
```

```python
pet = Pet('Dog','Beaglier','Bella','White',0b0,'Sally','123 Campus Dr, San Jose, CA 91231',date(2018,5,17))
print(pet)
print("\nPet Gender:")
pet.displayGender()
```

```python
dog = Dog('Beaglier','Bella','White',0b1)
dog.setBirthday(date(2016,1,2))
print(dog)
print("\nDog Gender:")
dog.displayGender()
```

```python
puppy = Puppy('Beaglier','Bella','White',0b1)
puppy.setBirthday(date(2016,1,2))
print(puppy)
print()
print(puppy.play())
```

# Pet Class II

```python
@dataclass
class Pet:
    breed: str 
    name: str
    color: str 
    gender: int 
    species: str 
    owner: str 
    address: str 
    birthday: datetime
    
    def setSpecies(self,new_species):
        self.species = new_species
    
    def setBreed(self,new_breed):
        self.breed = new_breed

    def setName(self,new_name):
        self.name = new_name
        
    def setColor(self,new_color):
        self.color = new_color
        
    def setGender(self,new_gender):
        self.gender = new_gender
        
    def setOwner(self,new_owner):
        self.owner = new_owner
        
    def setAddress(self,new_address):
        self.address = new_address
        
    def setBirthday(self,new_birthday):
        if type(new_birthday) == date:
            self.birthday = new_birthday
            
    def displayGender(self):
        if self.gender == 1:
            print('Male')
        elif self.gender == 0:
            print('Female')
        else:
            print('Unkonwn Gender')
    
    def getAge(self):
        try:
            now = date.today()
            return (now - self.birthday).days//365
        except TypeError:
            return None
    
    def displayAge(self):
        if self.getAge() == None:
            print("Unknow Age")
        else:
            print(self.name+" is "+str(self.getAge())+" year old.")
        
@dataclass
class Dog(Pet):
    species: str = 'Canis Familiaris'
    owner: str = None
    address: str = None
    birthday: datetime = None

    def getAge(self):
        try:
            now = date.today()
            return (now - self.birthday).days
        except TypeError:
            return None

    def displayAge(self):
        if self.getAge() == None:
            print("Unknow Age")
        else:
            print(self.name+" is "+str(self.getAge())+" day old.")

@dataclass
class Puppy(Dog):

    def play(self):
        return "(___()'`;\n      /,    /`\n      \\\\\"--\\\\"
    
    def getAge(self):
        try:
            now = date.today()
            return (now - self.birthday).days//30
        except TypeError:
            return None

    def displayAge(self):
        if self.getAge() == None:
            print("Unknow Age")
        else:
            print(self.name+" is "+str(self.getAge())+" month old.")
```

```python
pet = Pet('Beaglier','Bella','White',0b1,'Dog','Sally','123 Campus Dr, San Jose, CA 91231',date(2018,5,17))
print(pet)
print()
pet.displayAge()
print()
print("Address is "+pet.address)
print("Owner is "+pet.owner)
pet.setOwner('Mary')
print('Change the owener to '+pet.owner)
print('\nGender: ')
pet.displayGender()
```

```python
dog = Dog('Beaglier','Bella','White',0b0)
print(dog)
print()
dog.setBirthday(date(2016,1,2))
dog.displayAge()
print()
dog.setColor('Brown')
print('Change the color to '+dog.color)
print()
print('Gender: ')
dog.displayGender()
```
